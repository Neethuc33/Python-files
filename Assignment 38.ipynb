```python
image = Image.open('dogs.jpeg')
```

```python
from matplotlib import pyplot as plot
%matplotlib inline
plot.imshow(image)
```

```python
import numpy as np
image_np = np.asarray(image)
```

```python
image_np.shape
image_np = image_np.reshape((185*272,3))
image_np.shape
```

```python
from sklearn.cluster import KMeans 
wcss = []
for i in range(1,11):
    kmeans = KMeans (n_clusters =  i , init = 'k-means++')
    kmeans.fit(image_np)
    wcss.append (kmeans.inertia_)
wcss
```

```python
plt.plot (range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel ('wcss')
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(image_np)
```

```python
labels = kmeans.predict(image_np)
labels = labels.reshape((185,272))
plt.imshow(labels)
```

```python
centers = [centers.astype(int) for centers in kmeans.cluster_centers_]
blob = np.ones((100,100,3))
blob1 = blob * centers[0]
plt.imshow(blob1.astype(np.uint8))
```

```python
blob2 = blob * centers[1]
plt.imshow(blob2.astype(np.uint8))
```
